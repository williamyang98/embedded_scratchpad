```python
import numpy as np
import math
```

```python
# Op amp triangle integrator equations

# Vi-V+ = V+-(-Vcc)
# 2V+ = Vi-Vcc
# V+ = Vi/2 - Vcc/2

# KCL on V-
# When MOSFET is OFF
# Ic = (Vi-V+)/R
# Ic = (Vi-Vi/2 + Vcc/2)/R 
# Ic = 1/R * (Vi/2 + Vcc/2)

# When MOSFET is ON
# Ic = (Vi-V+)/R + (-Vcc-V+)/(R/2)
# Ic = (Vi-Vi/2+Vcc/2)/R + (-Vcc-Vi/2+Vcc/2)/(R/2)
# Ic = Vi/2R + Vcc/2R - Vcc/R - Vi/R
# Ic = -Vi/2R - Vcc/2R
# Ic = -1/R * (Vi/2 + Vcc/2)


# Ic = ±1/2R*(Vi+Vcc)

# C = Q/V
# V = Q/C = Ic*t/C
# dV/dt = Ic/C
# dV/dt = 1/2RC * (Vi+Vcc)
# SR = 1/2RC * (Vi+Vcc) 

# For a schmitt trigger window of ±Vth
# Vth = ±Vcc * R2/(R1+R2)


# SR * T = 4Vth (Needs to go up and down to generate one period)
# T = 4Vth/SR
# f = 1/T
# f = SR/4Vth

# f = 1/(8*RC*Vth) * (Vi+Vcc)
# Therefore f is proportional to Vi
# df/dVi = 1/(8*RC*Vth)
```

```python
R = 10e3
#C = 10e-9
C = 1e-9

Vcc = 2.5
Vth = Vcc * 1/10

SR = 1/(2*R*C) * (Vcc)
f = SR/(4*Vth)

df_dVi = 1/(8*R*C*Vth)

print(f"SR     = {SR*1e-6:.2f}V/us")
print(f"Vth    = {Vth:.2f}V")
print(f"f_0    = {f/1000:.2f}kHz")
print(f"df/dVi = {df_dVi/1000:.2f}kHz/V")

# calculate the max frequency considering the propagation delay
Td = 200e-9
fd = 1/Td * 1/5
# When considering the rise time as well
print(f"fd = {fd/1000:.2f}kHz")
```

```python
# PLL transfer response
dw_dVi = df_dVi*2*np.pi

# G2(s) = Avco / s
# G1(s) = 1/(1+sRC)

# G(s) = Avco/s(1+sRC)

# Feedback loop
# E(s) = X(s)-Y(s)
# Y(s) = G(s)E(s)
# Y(s) = G(s)X(s) - G(s)Y(s)
# Y(s)/X(s) = G(s)/[1+G(s)]

# H(s) = Avco/[s(1+sRC) + Avco]
# H(s) = Avco/(s^2 (RC) + s + Avco)
# H(s) = Avco/RC * 1/(s^2 + s/RC + Avco/RC)

# X(s) = k/s
# Y(s) = k * Avco/RC * 1/(s^2 + s/RC + Avco/RC) * 1/s

Rf = 5e3
Cf = 1e-9

Avco = dw_dVi
wn = (Avco/(Rf*Cf))**0.5
# 2*DF*wn = 1/RC, wn/Q = 1/RC
# Q = wn*RC = (Avco*Rf*Cf)**0.5
Q = wn*Rf*Cf

f_lpf = 1/(2*np.pi*Rf*Cf)

print(f"Avco={Avco:.2f}")
print(f"Q={Q:.2f}")
print(f"fc={f_lpf/1000:.2f}kHz")
```

```python
# Error value after LPF 
# F(s)G2(s) = Y(s)
# F(s) = Y(s)/G2(s)
# F(s)/X(s) = H(s)/G2(s)
# H2(s) = s/Avco * Avco/RC * 1/(s^2 + s/RC + Avco/RC)
# H2(s) = s/RC * 1/(s^2 + s/RC + Avco/RC)

# For X(s) = k/s
# F(s) = k/RC * 1/(s^2 + s/RC + Avco/RC)
# F(0) = k/RC * RC/Avco 

# Our steady state control voltage into the VCO
# F(0) = k/Avco
#f_target = 1/(50e-6)
f_target = 125e3
N_counter = 1

# our N-bit counter reduces the gain of the VCO

k = f_target*2*np.pi
ctrl_voltage = k/(Avco/N_counter)
print(f"ctrl_voltage={ctrl_voltage:.2f}V")
```
